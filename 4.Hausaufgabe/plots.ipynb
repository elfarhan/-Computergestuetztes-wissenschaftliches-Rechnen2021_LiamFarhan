```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu_1=np.loadtxt("mu1_values", delimiter=",")
mu_2=np.loadtxt("mu2_values", delimiter=",")
mu_3=np.loadtxt("mu3_values", delimiter=",")
data=np.loadtxt("x1000vsmu",delimiter=",")
```

```python
x0=mu_1[1][0]
def n(t):
    return x0*np.exp(r0*t)/(1+x0*K*(np.exp(r0*t)-1));
K=1
r0=1
def x(n,mu):
    r0delta_t=4*mu-1
    return n*K*r0delta_t/(1+r0delta_t);
def muu(t):
    return (1+t*r0)/4 ;
t=np.linspace(0,170,1000)
```

```python
plt.plot(mu_1[:,0], mu_1[:,1],label="$\mu_1=0,4$")
plt.plot(mu_2[:55,0], mu_2[:55,1],label="$\mu_2=0,74$")
plt.plot(mu_3[:22,0], mu_3[:22,1],label="$\mu_3=0,77$")
plt.plot(muu(t),x(n(t),muu(t)),label="x(t)")
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.legend()
plt.savefig("a3.png")
```
