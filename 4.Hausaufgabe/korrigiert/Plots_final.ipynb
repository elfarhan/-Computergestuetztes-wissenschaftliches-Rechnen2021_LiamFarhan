```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu_1=np.loadtxt("mu1_values", delimiter=",")
mu_2=np.loadtxt("mu2_values", delimiter=",")
mu_3=np.loadtxt("mu3_values", delimiter=",")
data=np.loadtxt("x1000vsmu",delimiter=",")
```

```python
def n(x,u):
    return 4*u*x/(4*u-1)
def n_analytisch(t,n_0,r_0,K):
    return (n_0*np.exp(r_0*t))/(1+K*n_0*(np.exp(r_0*t)-1))
def r(u):
    return (4*u-1)/u
```

```python
u=0.4
t=np.linspace(0,100*u,1000)
plt.plot(t,n_analytisch(t,n(0.1,0.4),r(u),1),label="analytisch")
plt.plot(mu_1[:,0],n(mu_1[:,1],u),label="$\mu=0,4$")
plt.xlabel("t")
plt.ylabel("$n(t)$")
plt.title("$\mu=0,4$")
plt.legend()
plt.savefig("mu=0,4.png")
```

```python
u=0.74
t=np.linspace(0,103*u,1000)
plt.plot(t,n_analytisch(t,n(0.1,0.4),r(u),1),label="analytisch")
plt.plot(mu_2[:,0],n(mu_2[:,1],u),label="$\mu=0,4$")
plt.xlabel("t")
plt.ylabel("$n(t)$")
plt.title("$\mu=0,74$")

plt.legend()
plt.savefig("mu=0,74.png")
```
