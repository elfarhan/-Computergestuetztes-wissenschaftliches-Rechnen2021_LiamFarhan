```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data3 = np.loadtxt("A3_values", delimiter=",")
```

```python
plt.plot(data3[:,0], data3[:,1])
plt.title("erf_simpson")
plt.xlabel("x")
plt.ylabel("$\int_0^x e^{(-y^2)}dy$")
plt.show()
```

```python
plt.plot(data3[:,0], data3[:,2])
plt.title("erf_midpoint")
plt.xlabel("x")
plt.ylabel("$\int_0^x e^{(-y^2)}dy$")
plt.show()
```

```python
data4 = np.loadtxt("A4_values", delimiter=",")
```

```python
plt.plot(np.log(data4[:,0]), np.log(abs(data4[:,1]-np.sinh(1))))
plt.title("Fehler der simpson Methode")
plt.xlabel("log(dx)")
plt.ylabel("$\log(I-cosh_{simpson}(dx))$")
plt.show()
```

```python
plt.plot(np.log(data4[:,0]), np.log(abs(data4[:,2]-np.sinh(1))))
plt.title("Fehler der midpoint Methode")
plt.xlabel("log(dx)")
plt.ylabel("$\log(I-cosh_{midpoint}(dx))$")
plt.show()
```

```python
#Die Abweichung zwischen den nummerischen Methoden und analytischem Wert des Integrals ist sehr klein und 
# unterliegt groe Schwankungen.
# Man wurde erwarten, dass log(I-cosh_simpson) =4 log(dx)
# und log(I-cosh_midpoint) = 2log(dx)
```
