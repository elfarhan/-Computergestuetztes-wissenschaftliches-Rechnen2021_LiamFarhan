```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt("A3_values", delimiter=",")
x=np.linspace(-0.02,2.02,1000)
y=x*0 
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(data[:,0],data[:,1])
plt.plot(x,y)
plt.xlabel("$\lambda$")
plt.ylabel("$p(\lambda)$")
plt.savefig("charPolynom.png")
```
