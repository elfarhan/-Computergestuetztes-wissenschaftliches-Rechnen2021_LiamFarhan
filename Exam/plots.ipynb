```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = np.loadtxt("A1_values", delimiter=",")
p = data1[:,0]
rho_mittel = data1[:,1]
rho_stdev = data1[:,2]
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(p[:501], rho_mittel[:501] ,'.', label="$d=1$\n$L=10$\n$N=1000$\n$t=500$")
plt.xlabel("p")
plt.ylabel("$\\rho$")
plt.legend()
plt.savefig("A1_d1.png")
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(p[501:1002], rho_mittel[501:1002] ,'.', label="$d=3$\n$L=10$ \n$N=1000$\n$t=500$")
plt.xlabel("p")
plt.ylabel("$\\rho$")
plt.legend()
plt.savefig("A1_d3.png")
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(p[1002:], rho_mittel[1002:] ,'.', label="$d=2$\n$L=10$ \n$N=1000$\n$t=500$")
plt.xlabel("p")
plt.ylabel("$\\rho$")
plt.legend()
plt.savefig("A1_d2.png")
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(p[:501], rho_stdev[:501] ,'.', label="$d=1$\n$L=10$ \n$N=1000$\n$t=500$")
plt.xlabel("p")
plt.ylabel("$\\sigma$")
plt.legend()
plt.savefig("A1_d1_stdev.png")
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(p[501:1002], rho_stdev[501:1002] ,'.', label="$d=3$\n$L=10$ \n$N=1000$\n$t=500$")
plt.xlabel("p")
plt.ylabel("$\\sigma$")
plt.legend()
plt.savefig("A1_d3_stdev.png")
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(p[1002:], rho_stdev[1002:] ,'.', label="$d=2$\n$L=10$ \n$N=1000$\n$t=500$")
plt.xlabel("p")
plt.ylabel("$\\sigma$")
plt.legend()
plt.savefig("A1_d2_stdev.png")
```

```python
data2 = np.loadtxt("A2_values", delimiter=",")
N = data2[:,0]
rho_mittel2 = data2[:,1]
rho_stdev2 = data2[:,2]
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(N, rho_mittel2 ,'.', label="$d=3$\n$L=10$\n$t=500$\n$p=0.05$")
plt.xlabel("N")
plt.ylabel("$\\rho$")
plt.legend()
plt.savefig("A2_rho.png")
```

```python

plt.figure(figsize=(7,4),dpi=300)

plt.grid(lw=0.7)
plt.yscale('log')
plt.xscale('log')
plt.plot(N, rho_stdev2 ,'.', label="$d=3$\n$L=10$\n$t=500$\n$p=0.05$")
plt.xlabel("N")
plt.ylabel("$\\sigma$")

plt.legend()
plt.savefig("A2_rho_stdev.png")

```
