```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
l = np.loadtxt("A1_Eigenvalues")
n = np.array([i for i in range(20)])
def w(n):
    return n+1/2
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(n,l, label="nummerisch bestimmt")
plt.plot(n,w(n),label="Analytischer Wert")
plt.xlabel("$n$-ter Eigenwert")
plt.ylabel("$\lambda$")
plt.legend()
plt.savefig("EigenwerteA1.png")
```

```python
N=500
psi = np.loadtxt("A1_Eigenvectors")
psi_1 = abs(psi[0:500])
psi_2 = abs(psi[500:1000])
psi_3 = abs(psi[1000:1500])
x = [i*2*5/500 -5 for i in range(500)]
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(x,psi_1,'.', label="$\psi_0$")
plt.plot(x,psi_2,'.', label="$\psi_1$")
plt.plot(x,psi_3,'.', label="$\psi_2$")
plt.xlabel("x")
plt.ylabel("$\psi(x)$")
plt.legend()
plt.savefig("Eigenzustnde.png")
```

```python
data = np.loadtxt("A2_Eigenvalues", delimiter=",")
delta_x = data[:,0]
ld1 = data[:,1]
ld2 = data[:,2]
ld3 = data[:,3]
ld4 = data[:,4]
resdue1 = []
resdue2 = []
resdue3 = []
resdue4 = []

for i in range(len(delta_x)-1):
    resdue1.append(abs((ld1[i+1]-ld1[i])/ld1[i+1]))
    resdue2.append(abs((ld2[i+1]-ld2[i])/ld2[i+1]))
    resdue3.append(abs((ld3[i+1]-ld3[i])/ld3[i+1]))
    resdue4.append(abs((ld4[i+1]-ld4[i])/ld4[i+1]))
```

```python
plt.figure(figsize=(7,4),dpi=300)
plt.grid(lw=0.7)
plt.plot(delta_x[:-1],resdue1,'.', label="$n=0$")
plt.plot(delta_x[:-1],resdue2,'.', label="$n=1$")
plt.plot(delta_x[:-1],resdue3,'.', label="$n=2$")
plt.plot(delta_x[:-1] ,resdue4,'.', label="$n=3$")
plt.xlabel("$\Delta x$")
plt.ylabel("$|E_\\alpha(\Delta x_i)-E_\\alpha(\Delta x_{i-1})/E_\\alpha(\Delta x_i)|$")
plt.legend()
plt.savefig("Resediuen.png")
```
